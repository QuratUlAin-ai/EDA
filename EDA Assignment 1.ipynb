There are some movies which are directed by multiple directors. Hence multiple names of directors are given in comma separated format.
We will explode the director column as well. It will create many duplicate records in originaltable hence we
created separate table for directors.

```python
#dropna() is used to remove any rows from dir_tb that contain missing values (NaN).
#inplace=True means the operation modifies dir_tb directly, so no new DataFrame is returned.
#.apply(lambda x: x.split(',')) applies a lambda function to each element of the 'director' column.
#lambda x: x.split(',') is a function that splits a string x on commas, returning a list of strings.
#This means that if a cell in the 'director' column contains a string with comma-separated values (e.g., "John Doe, Jane Smith"), it will be split into a list ["John Doe", "Jane Smith"]
dir_tb = data[['show_id' , 'type' , 'director']]
dir_tb.dropna(inplace = True)
dir_tb['director'] = dir_tb['director'].apply(lambda x : x.split(','))
dir_tb
```

```python
dir_tb = dir_tb.explode('director')
```

```python
dir_tb['director'] = dir_tb['director'].str.strip()
```

```python
dir_tb.director.apply(lambda x : True if len(x) == 0 else False).value_counts()
```

```python
dir_tb
```

```python
dir_tb['director'].nunique()
```

Total movies and tv shows directed by each director

```python
#.reset_index(): After performing the grouping and counting, reset_index() converts the result from a Series to a
#DataFrame and resets the index. The resulting DataFrame x will have columns 'director', 'type', and 'show_id', where 'show_id' contains
# the count of shows for each combination of 'director' and 'type'.
#ascending=False specifies that the DataFrame should be sorted in descending order, so directors with the most movies appear at the top
x = dir_tb.groupby(['director' , 'type'])['show_id'].count().reset_index()
x.pivot(index= ['director'] , columns = 'type' , values = 'show_id').sort_values('Movie' ,ascending = False)
```

#  4. Visual Analysis - Univariate & Bivariate

Distribution of content across the different types (Movies/TV Shows) using
matplotlib pie chart

```python
types = data.type.value_counts()
plt.pie(types,  labels=types.index, autopct='%1.1f%%' , colors = ['g', 'b'])
plt.title('Total_Movies and TV Shows')
plt.show()
```

General visual analysis of null values in each column

```python
# Counting null values per column and plotting
null_counts = data.isnull().sum()

plt.figure(figsize=(12, 6))
null_counts.plot(kind='bar', color='skyblue')
plt.title("Number of Missing Values in Each Column")
plt.ylabel("Number of Missing Values")
plt.xlabel("Columns")
plt.show()
```

Seaborn box plot for each numeric column to identify any outliers.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Select only numeric columns
numeric_columns = data.select_dtypes(include=['float64', 'int64']).columns

# Plotting box plots for each numeric column
plt.figure(figsize=(15, 8))
for i, column in enumerate(numeric_columns, 1):
    plt.subplot(len(numeric_columns), 1, i)
    sns.boxplot(x=data[column], color='lightblue')
    plt.title(f"Box Plot of {column}")

plt.tight_layout()
plt.show()
```
