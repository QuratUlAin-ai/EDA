```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Simplest plot

```python
x = np.arange(11)

plt.plot(x, x**2)
```

### Let's add more features to the figure and the plot above

```python
x = np.arange(11)

plt.figure(figsize = (10,5), edgecolor = 'b', frameon = True, linewidth = 5, facecolor = 'w') #all the changes to background figure/canvas are made here

plt.plot(x, x**2, marker = "v", color = "k", linestyle = "--") #all changes to the plot are made here
plt.plot(x, -x, "bo--") #using a format string instead of separate keyword arguments for color, marker and linestyle
plt.plot(x, x, "y")


plt.title("My First Graph", color = 'r', size = 15, loc = 'right', family = 'serif') #loc tells us location and family tells us font
plt.xlabel("x")
plt.ylabel("x**2")
plt.xlim(-20, 20)
plt.ylim(-10, 120)

```

### Plotting on the same figure

```python
#plotting both plots on the same figure.
#we initiate ONE figure and then we have two plot statements on the same figure

x = np.arange(-10, 11)

plt.figure()
plt.title("first plot")

#if you add labels to each plot using the label keyword argument, you can also add a legend. without individual labels, a legend won't work
plt.plot(x, x**2, "k--", label = "x vs x sq")
plt.plot(x, -x, "ro", label = "x vs -x")


plt.legend(loc = "upper center")
```

### Plotting on the different figures

```python
#to plot on separate figures, initiate a figure instance each time before the plot statement

x = np.arange(-10, 11)

plt.figure()
plt.plot(x, x**2, "k--", label = "x vs x sq")

plt.figure()
plt.plot(x, -x, "ro", label = "x vs -x")
```

### Storing your plot and axes in variables to plot on a figure from a different cell

```python
#you can store your figure and axes in variables. this way, you can call on them/ plot them from a separate cell too.

x = np.arange(-10, 11)

fig = plt.figure()
axes = plt.axes()


axes.plot(x, -x, "r")
```

```python
#plotting from a separate cell on the plot above

axes.plot(x, x**2, "b")
fig
```

## plt.subplots()
can be used to generate one plot or multiple subplots on the same figure <br>
this method returns both figure and axes to you, so you don't need to initiate them separately like above <br>
often used to plot single graphs because it returns both figure and axes

##### NOTE:
with plt.plot() your commands are plt.title(), plt.xlabel(), plt.ylim() etc. <br>
when you store your axes in a variable, and call on axes, then the commands become ax.set_title(), ax.set_xlabel(), ax.set_ylim() etc.


```python
#creating a single plot with plt.subplots(). simply skip the first two arguments for rows and columns

x = np.arange(-10,11)
fig, axes = plt.subplots(figsize= (10,8), linewidth = 2, edgecolor = 'b', facecolor = "w") #all the figure changes are made here now

axes.plot(x, -x, "r")
axes.set_title("x vs -x", color = 'w', loc = 'center', size = 15)
axes.set_xlabel("-x", color = 'w', size = 15) #you can change xlabel/ylabel size too
axes.set_ylabel("x", color = 'w', size = 15)
```

```python
#you can create more than one subplots in the same figure. this time the first two arguments will be rows, columns

x = np.arange(-10,11)
fig, axes = plt.subplots(2,3, figsize= (10,8), edgecolor = "b", linewidth = 2)



axes[0,2].plot(x,-x, "r")
axes[0,2].set_title("third plot")
axes[0,2].set_xlabel("-x", size = 15)
axes[0,2].set_ylabel("x", size = 15)

axes[1,1].plot(x, x**2, "ko")
axes[1,1].set_title("fifth plot")
```

```python
#you can also unpack the subplots like this and assign each to its own variable

x = np.arange(-10,11)
fig, ((ax1, ax2, ax3), (ax4,ax5,ax6)) = plt.subplots(2,3, figsize= (10,8), edgecolor = "b", linewidth = 2)


ax2.plot(x,-x, "r")
ax2.set_title("y = -x", color = "r", size = 10)
ax2.set_xlabel("-x")
ax2.set_ylabel("y")
ax2.set_xlim(-15, 15)
ax2.set_ylim (-15, 15)

ax5.plot(x, x**2, "g")
ax5.set_title("y = x**2", color = "b", size = 10)
ax5.set_xlabel("x")
ax5.set_ylabel("y")







#NOTE:
#as you can see, the y_label of ax6 overlaps with the ax5 plot
#For EACH subplot, the default coordinates of x_label and y_label are (0, 0.5) where (0,0) is the bottom left of the subplot and (0.5, 0.5) is the center
#lets see how we can see the coordinates of our y_label.

#to get the position coordinates, uncomment the following line:
# ax6.set_ylabel("y").get_position()


#simply adjust these coordinates with axes.yaxis.set_label_coords() or axes.xaxis.set_label_coords() like below.
#uncomment the following two lines and run the cell again. ax2 xlabel and ylabel positions have changed

# ax2.yaxis.set_label_coords(-0.1, 0.5)
# ax2.xaxis.set_label_coords(0.5,-0.08)
```

## DIFFERENT TYPES OF PLOTS

### SCATTER PLOT

```python
x = np.arange(1,5)
y = np.arange(1,5)

sizes = np.array([10,20,30,40])
colors = np.random.rand(4)

plt.scatter(x, y, s = sizes * 10, c = colors, cmap = 'RdBu_r', alpha = 1, edgecolors = "r")
plt.title("scatter plot")


#plt.scatter is taking in the following
# 1) a list of 4 x values
# 2) a list of 4 y values
# 3) a list of 4 numbers for s which determines size of each point
# 4) a list of 4 numbers for colors (that we randomly generated and stored in the variable colors)
# 5) a single value for alpha, which is the transparency
# 6) a single value for cmap. cmap is the color theme matplotlib will use to map your colors array (which contains numbers here) to colors

# it will now cycle through the lists, grabbing x[0], y[0] as the point to plot, will make area[0] its size and color[0] its color
# then it'll do the same for the next set of points i.e. x[1],y[1],area[1],color[1] and so on
```

### BAR GRAPH

```python
countries = {"Singapore" : 100000, "Pakistan": 4000, "China": 50000, "Indonesia": 10000}

plt.bar(countries.keys(), countries.values(), color = 'g', width = 0.5)
```

```python
#to set a different color for each individual bar, you can store the plot in a variable and treat each bar like a separate subplot
#then you can use a loop to plot on bars[0], bars[1], bars[2] etc

countries = {"Singapore" : 100000, "Pakistan": 4000, "China": 50000, "Indonesia": 10000}
bars = plt.bar(countries.keys(), countries.values(), width = 0.5)


colors = ['r', 'b', 'y', 'k']
for i in range(len(bars)):
    bars[i].set_color(colors[i])
```

### PIE CHART

```python
#pie chart

x = ["a", "b", "c"]

y = [70, 60, 54]

colors = ["k", "r", "b"]

plt.pie(y, labels = x, colors = colors)
```

```python
#you can plot a pie chart with a dictionary too

countries = {"Singapore" : 100000, "Pakistan": 4000, "China": 50000, "Indonesia": 10000}
colors = ["g", "k", "b", "y"]

plt.pie(countries.values(), labels = countries.keys(), colors = colors)
```

### HISTOGRAMS

```python
#plotting a histogram using a thousand random values from the normal distribution

values = np.random.randn(1000)

plt.hist(values, bins = 6, edgecolor = 'k', alpha = 0.5, color = "g")
```

### BOXPLOT

```python
values = np.concatenate([np.random.randn(10), np.array([-15, -10, 10, 15])]) #Concatenating an array of random numbers from the normal dist and an array of 4 outliers

plt.boxplot(values)
```

## subplot2grid

```python
ax1 = plt.subplot2grid((3,3), (0,0), rowspan = 2, colspan = 3)
ax2 = plt.subplot2grid((3,3), (2,0))
ax3 = plt.subplot2grid((3,3), (2,1), colspan = 2 )


#first argument (3,3) is a tuple specifying the arrangement of subplots. this means 3 rows and 3 columns
#second argment (0,0) is a tuple of position coordinates for the subplot within the larger arrangement
# rowspan and colspan tells how many rows and columns you want the subplot to span
```
